```python
X = data[["memory_usage_pct", "rolling_std_24h", "growth_rate", "acceleration"]]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python

from sklearn.ensemble import IsolationForest

model = IsolationForest(
    contamination="auto",
    random_state=42
)

model.fit(X_scaled)
```

```python
score = model.decision_function(X_scaled)
data["anomaly_score"] = -score
```

```python

import matplotlib.pyplot as plt 
import seaborn as sns

data['ts'] = pd.to_datetime(data['ts'], format='mixed').dt.tz_localize(None)
```

```python
sns.lineplot(x=data.ts, y=data.memory_usage_pct)
sns.jointplot(data=data, x=data.ts, y=data.memory_usage_pct, hue=data.anomaly_score)
```

```python
g = sns.jointplot(data=data, x=data.ts, y=data.memory_usage_pct, hue=data.anomaly_score)
sns.lineplot(data=data, x=data.ts, y=data.memory_usage_pct, ax=g.ax_joint, color='purple', alpha=0.5, zorder=0)
plt.show()
```
