```python
import pandas as pd
import numpy as np

from pathlib import Path

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 20)
```

```python
# Load CPU Datasets
cpu1 = pd.read_csv(
    r"C:\Users\Avilasha\Desktop\CPU_Predictive_Maintenance\CPU\data\Raw\Datacenter Datas 11.06.2026 to 18.06.2026\SERVER WISE DATA\host_metrics_host1.csv"
)
server1_resource = pd.read_csv(
    r"C:\Users\Avilasha\Desktop\CPU_Predictive_Maintenance\CPU\data\Raw\Datacenter Datas 11.06.2026 to 18.06.2026\SERVER WISE DATA\ilo_server_metrics_iloserver1.csv",
    usecols=[
        "ts",
        "ilo_server_id",
        "inlet_temp_c",
        "cpu_temp_c",
        "processor_count"
    ]
)

cpu2 = pd.read_csv(
    r"C:\Users\Avilasha\Desktop\CPU_Predictive_Maintenance\CPU\data\Raw\Datacenter Datas 11.06.2026 to 18.06.2026\SERVER WISE DATA\host_metrics_host2.csv"
)
server2_resource = pd.read_csv(
    r"C:\Users\Avilasha\Desktop\CPU_Predictive_Maintenance\CPU\data\Raw\Datacenter Datas 11.06.2026 to 18.06.2026\SERVER WISE DATA\ilo_server_metrics_iloserver2.csv",
    usecols=[
        "ts",
        "ilo_server_id",
        "inlet_temp_c",
        "cpu_temp_c",
        "processor_count"
    ]
)

cpu3 = pd.read_csv(
    r"C:\Users\Avilasha\Desktop\CPU_Predictive_Maintenance\CPU\data\Raw\Datacenter Datas 11.06.2026 to 18.06.2026\SERVER WISE DATA\host_metrics_host3.csv"
)
server3_resource = pd.read_csv(
    r"C:\Users\Avilasha\Desktop\CPU_Predictive_Maintenance\CPU\data\Raw\Datacenter Datas 11.06.2026 to 18.06.2026\SERVER WISE DATA\ilo_server_metrics_iloserver3.csv",
    usecols=[
        "ts",
        "ilo_server_id",
        "inlet_temp_c",
        "cpu_temp_c",
        "processor_count"
    ]
)
```

```python
# inspect CPU Datasets
print("CPU1")
display(cpu1.head())
print("server1_resource")
display(server1_resource.head())

print("CPU2")
display(cpu2.head())
print("server2_resource")
display(server2_resource.head())

print("CPU3")
display(cpu3.head())
print("server3_resource")
display(server3_resource.head())
```

```python
# Check column names
print(cpu1.columns)
print(cpu2.columns)
print(cpu3.columns)
print(server1_resource.columns)
print(server2_resource.columns)
print(server3_resource.columns)
```

```python
# convert timestamp columns
cpu1["ts"] = pd.to_datetime(
    cpu1["ts"],
    format="mixed",
    utc=True
)

cpu2["ts"] = pd.to_datetime(
    cpu2["ts"],
    format="mixed",
    utc=True
)

cpu3["ts"] = pd.to_datetime(
    cpu3["ts"],
    format="mixed",
    utc=True
)

server1_resource["ts"] = pd.to_datetime(
    server1_resource["ts"],
    format="mixed",
    utc=True
)

server2_resource["ts"] = pd.to_datetime(
    server2_resource["ts"],
    format="mixed",
    utc=True
)

server3_resource["ts"] = pd.to_datetime(
    server3_resource["ts"],
    format="mixed",
    utc=True
)
```

```python
# verify dataset size
print(cpu1.shape)
print(cpu2.shape)
print(cpu3.shape)

print("\nTotal rows expected:")
print(len(cpu1) + len(cpu2) + len(cpu3))
```

```python
HOST_MAP = {
    1: "10.10.10.65",
    2: "10.10.10.150",
    3: "10.10.10.2"
}
cpu1["hostName"] = cpu1["host_id"].map(HOST_MAP)

cpu2["hostName"] = cpu2["host_id"].map(HOST_MAP)

cpu3["hostName"] = cpu3["host_id"].map(HOST_MAP)
```

```python
# combining all CPU Datasets
server1 = pd.merge(
    cpu1,
    server1_resource,
    left_on=["ts", "host_id"],
    right_on=["ts", "ilo_server_id"],
    how="inner"
)

server2 = pd.merge(
    cpu2,
    server2_resource,
    left_on=["ts", "host_id"],
    right_on=["ts", "ilo_server_id"],
    how="inner"
)

server3 = pd.merge(
    cpu3,
    server3_resource,
    left_on=["ts", "host_id"],
    right_on=["ts", "ilo_server_id"],
    how="inner"
)

print(server1.shape)
print(server2.shape)
print(server3.shape)
```

```python
cpu_master_new = pd.concat(
    [server1, server2, server3],
    ignore_index=True
)

HOST_MAP = {
    1: "10.10.10.65",
    2: "10.10.10.150",
    3: "10.10.10.2"
}

cpu_master_new["hostName"] = (
    cpu_master_new["host_id"]
    .map(HOST_MAP)
)
cpu_master_new.shape
```

```python
# timestamp range
print("Earliest Timestamp:")
print(cpu_master_new["ts"].min())

print("\nLatest Timestamp:")
print(cpu_master_new["ts"].max())
```

```python
# Sort dataset
cpu_master = cpu_master_new.sort_values(
    ["hostName", "ts"]
)
```

```python
# reset index
cpu_master = cpu_master_new.reset_index(
    drop=True
)
```

```python
# save processed dataset
Path("../data/Processed").mkdir(
    parents=True,
    exist_ok=True
)
cpu_master.to_csv(
    "../data/Processed/cpu_master_new.csv",
    index=False
)
```
