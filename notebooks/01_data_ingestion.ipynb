```python
import pandas as pd
import numpy as np

from pathlib import Path

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 20)
```

```python
# Load CPU Datasets
cpu1 = pd.read_csv(
    r"C:\Users\Avilasha\Desktop\CPU_Predictive_Maintenance\data\Raw\DataCenter Last month data\SERVER WISE DATA\CPU DATA\superadmin-cpu-custom (1).csv"
)

cpu2 = pd.read_csv(
    r"C:\Users\Avilasha\Desktop\CPU_Predictive_Maintenance\data\Raw\DataCenter Last month data\SERVER WISE DATA\CPU DATA\superadmin-cpu-custom (2).csv"
)

cpu3 = pd.read_csv(
    r"C:\Users\Avilasha\Desktop\CPU_Predictive_Maintenance\data\Raw\DataCenter Last month data\SERVER WISE DATA\CPU DATA\superadmin-cpu-custom.csv"
)
```

```python
# inspect CPU Datasets
print("CPU1")
display(cpu1.head())

print("CPU2")
display(cpu2.head())

print("CPU3")
display(cpu3.head())
```

```python
# Check column names
print(cpu1.columns)
print(cpu2.columns)
print(cpu3.columns)
```

```python
# convert timestamp columns
cpu1["timestamp"] = pd.to_datetime(cpu1["timestamp"])
cpu2["timestamp"] = pd.to_datetime(cpu2["timestamp"])
cpu3["timestamp"] = pd.to_datetime(cpu3["timestamp"])
```

```python
# verify dataset size
print(cpu1.shape)
print(cpu2.shape)
print(cpu3.shape)

print("\nTotal rows expected:")
print(len(cpu1) + len(cpu2) + len(cpu3))
```

```python
# combining all CPU Datasets
cpu_master = pd.concat(
    [cpu1, cpu2, cpu3],
    ignore_index=True
)
cpu_master.shape
```

```python
# inspect combined dataset
cpu_master.head()
# verify unique hosts
cpu_master["hostName"].unique()
```

```python
# CPU Dataset information
cpu_master.info()
```

```python
# missing values check
cpu_master.isnull().sum()
```

```python
# duplicate check
cpu_master.duplicated().sum()
```

```python
# timestamp range
print("Earliest Timestamp:")
print(cpu_master["timestamp"].min())

print("\nLatest Timestamp:")
print(cpu_master["timestamp"].max())
```

```python
# Sort dataset
cpu_master = cpu_master.sort_values(
    ["hostName", "timestamp"]
)
```

```python
# reset index
cpu_master = cpu_master.reset_index(
    drop=True
)
```

```python
# save processed dataset
Path("../data/Processed").mkdir(
    parents=True,
    exist_ok=True
)
cpu_master.to_csv(
    "../data/Processed/cpu_master.csv",
    index=False
)
```
